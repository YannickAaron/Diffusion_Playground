# Score Matching

Propensity score matching (PSM) is a statistical matching approach used in the statistical analysis of observational data that aims to quantify the effect of a treatment, policy, or other intervention by accounting for the factors that predict getting the treatment. PSM aims to eliminate the bias caused by confounding variables in an estimate of the treatment effect derived by merely comparing outcomes between units that got the treatment and those that did not.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_swiss_roll
from diffusionplayground.helper.plot import hdr_plot_style
hdr_plot_style()
# Sample a batch from the swiss roll
def sample_batch(size, noise=1.0):
    x, _= make_swiss_roll(size, noise=noise)
    return x[:, [0, 2]] / 10.0
# Plot it
data = sample_batch(10**4).T
plt.figure(figsize=(16, 12))
plt.scatter(*data, alpha=0.5, color='red', edgecolor='white', s=40)
```
